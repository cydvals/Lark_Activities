```python
for col in df.columns:
    if((df[col].dtype=='object') and df[col].isin(['yes','no']).any()):
        df[col]=df[col].map({'yes':1, 'no':0})
df.head()
```

```python
# changes the yes/no categorical data into binary (1/0)
```

```python
df=pd.get_dummies(df)
df    
```

```python
# Performing one hot encoding to convert categorical data 
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler(feature_range=(0,1))
DF = pd.DataFrame(min_max.fit_transform(df),columns = df.columns)
X=DF.drop(['passed'],axis=1)
y=DF['passed']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.25)
X_train.shape
```
